```python
#calculer le profit
def Calc_profit(L:list):
    p=0
    for i in L :
        p=p+int(D_Profit[int(i)])
    return p
```

```python
#Q1 :Proposer une heuristique qui donne une solution initiale réalisable
#Trier les objets par densité
#ajouter les "meuilleurs" objets selon la densité 
# arreter quen le sac est plein
def remplir_sac_init(Sac:bitboard,N_obj):#remplissage initiale du sac dénoté par le bitboard b: represente la solution S0 
    D_Densité={i:float(D_Profit[i])/float(D_Volume[i]) for i in range(N) } #dict indice: densité
    D_Densité_re={float(D_Profit[i])/float(D_Volume[i]):i for i in range(N) } #dict densité: indice
    L0=list(D_Densité.values())
    L0.sort()#conversion en liste triée des densités
    V=0
    Compteur=int(copy.deepcopy(N_obj))
    while V<=C:#le volume occupé < capacité
        #ajouter les objets au sac avec la plus haute densité 
        V=V+int(D_Volume[D_Densité_re[L0[(Compteur-1)]]]) #mettre à jour le volume du sac
        if V<=C:
            Sac.setBitHot(D_Densité_re[L0[Compteur-1]])#mettre l'objet dans le sac
        else :
            V=V-int(D_Volume[D_Densité_re[L0[Compteur-1]]])
            break
        L0.pop() #mettre à jour la liste de densité<->supprimer l'objet ajouté au sac
        Compteur-=1#décrementer le rang des densités à choisir
    Sac.volume=V
    Sac.profit=Calc_profit(Sac.get_squares_from_bb())
    return Sac #retourne le sac remplit
```

```python
#Q2:propser une heurestique à base de recherche locale
#décrire la structure de voisinage : switch i et j <-> distance de Hamming <=2 ou (plus de 2 choix de l'éxperimentateur en optimisant)
#décrire la stratégie d'exploration : la Descente la plus rapide si en trouve une solution dans le voisinage courant qui améliore
#on on prend cette solution et crée itérativement son propre voisinage jusqu'à atteindre un optimum qu'on espére etre global
#une comparaison avec la méthode exacte (Ciplex) ou avec les résultats d'un nombre assez important de départs peut trancher l'arrivé
```

```python
def hamming2(s1, s2):
    """Calculate the Hamming distance between two bit strings"""
    assert len(s1) == len(s2)
    return sum(c1 != c2 for c1, c2 in zip(s1, s2))
```

```python
def add_check(b:bitboard,Obj_index,Capacité):
        return (b.volume+int(D_Volume[Obj_index])<=Capacité)
#volume du sac avec objet < C  ,l'objet est acceptable est la solution est encore réalisable sinon return false
```

```python
def Construct_bb(L:list):
    y=bitboard()
    for i in set(L):
        y.setBitHot(i)
    return y
```

```python
def Construct_voisinage(b1):
    L_voisins=list()
    L=b1.get_squares_from_bb()
    for i in range(N):
        x=b1
        if i in L:
            x.clearBit(i)
        for j in range(N):
            if i==j:
                continue
            if (add_check(x,j,C)) :
                x.setBitHot(j)
                if (x.get_squares_from_bb() in L_voisins)==False:
                    L_voisins.append(x.get_squares_from_bb())         
    return L_voisins,[Construct_bb(i) for i in L_voisins]
```

```python
Construct_voisinage(b)
```

```python
def Descente_rap(Solution:bitboard):
    Srl=Solution
    Aml=True
    while Aml :
        Aml=False
        fmax=0
        for i in Construct_voisinage(Solution)[1] :
            if Calc_profit(i.get_squares_from_bb())>fmax:
                Smax=i
                fmax=Calc_profit(i.get_squares_from_bb())
        if fmax>Calc_profit(Srl.get_squares_from_bb()):
            Srl=i
            Aml=True
    return  Srl
```

```python
#Q3 proposer une heuristique constructive randomisée permettant de construire une solution initiale de maniére aléatoire pour le PSAD
def remplir_sac_init_alea(Sac:bitboard,N_obj):#remplissage initiale du sac dénoté par le bitboard b: represente la solution S0   
    Objs=list(range(N_obj))
    V=Sac.volume
    while V <= C :
        rand=random.choice(Objs)
        if add_check(Sac,int(rand),C) and len(Objs)>1: 
            Objs.remove(rand)  #eviter la reprise d'un objet une fois choisit
            V=V+int(D_Volume[rand]) #mettre à jour le volume du sac
        if V>C or len(Objs)<=1:
            V=V-int(D_Volume[rand]) #si l'objet qu'on d'ajouter est trop volumineux on doit le retirer
            break
        Sac.setBitHot(rand)#mettre l'objet dans le sac (ne sera pas executé si rand est trop volumineux)
    Sac.volume=V
    Sac.profit=Calc_profit(Sac.get_squares_from_bb())
    return Sac
```

```python
#Q4 proposer une métaheuristique GRASP pour le PSAD 
# l'heuristuqe stochastique ultérieur servira d'un point de départ suffisant en effet , on peut simuler les départs multiples
#on doit définir les voisinage avec lesquel se fait l'exploration de l'espace de recherche 
#avec le départ stochastique on peut éviter les opimums locaux pour la meme instance
e=bitboard()
e=remplir_sac_init_alea(e,N) 
Sopt=Descente_rap(e)
Sopt.get_squares_from_bb(),Sopt.profit
```

```python
#Q5 proposer une métaheuristique Tabou simple
    #on peut tenir compte d'une liste de longueur 3 contenant les solutions voisine réalisables 
    #et bloquer l'ajout d'une ancienne solution à la liste courante de voisin s'elle fait partie de la liste tabou
def Construct_voisinage_avec_tabou(b1):
    L=list()
    L.append(0)
    L.append(0)
    L.append(0)
    Compteur=0
    while Descente_rap(b1)!=b1 or Compteur==15: #on suppose ces 2critéres d'arret avec 15 étant un nombre d'itérations fixé par l'expérimentateur 
        z=copy.deepcopy(b1)
        New =Descente_rap(z)
        L[Compteur%3]=z
        Compteur +=1
        if New in Construct_voisinage(b1)[1]:
            New_Voisinage=Construct_voisinage(b1)[1].remove(New)
    return New_Voisinage
```

```python
def Descente_rap_avec_tabou(Solution:bitboard):
    Srl=Solution
    Aml=True
    while Aml :
        Aml=False
        fmax=0
        for i in Construct_voisinage_avec_tabou(Solution)[1] :
            if Calc_profit(i.get_squares_from_bb())>fmax:
                Smax=i
                fmax=Calc_profit(i.get_squares_from_bb())
        if fmax>Calc_profit(Srl.get_squares_from_bb()):
            Srl=i
            Aml=True
    return  Srl
```
