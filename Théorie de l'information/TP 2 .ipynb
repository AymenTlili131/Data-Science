```python
#Q7
X1=stats.norm.rvs(size=1000)
X2=stats.uniform.rvs(size=1000)
X3=stats.triang.rvs(c=0.5,size=1000)
X4=stats.expon.rvs(size=1000)
```

```python
#Q8
abs1=np.linspace(-4,4, 1000)
abs2=np.linspace(-2,2, 1000)
abs3=np.linspace(-1,2, 1000)
abs4=np.linspace(-1,5, 1000)
```

```python
pdf1=stats.norm.pdf(abs1,loc=0,scale=1)
cdf1=stats.norm.cdf(abs1,loc=0,scale=1)
pdf2=stats.uniform.pdf(abs2,loc=0,scale=1)
cdf2=stats.uniform.cdf(abs2,loc=0,scale=1)
pdf3=stats.triang.pdf(abs3,c=0.5,loc=0,scale=1)
cdf3=stats.triang.cdf(abs3,c=0.5,loc=0,scale=1)
pdf4=stats.expon.pdf(abs4,loc=0,scale=1)
cdf4=stats.expon.cdf(abs4,loc=0,scale=1)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.hist(X1,density=True, edgecolor="black",label="histogram",bins=25)
plt.plot(abs1,pdf1,label="pdf")
plt.plot(abs1,cdf1,label="cdf")
plt.legend()
plt.title("Normale")
plt.subplot(2,2,2)
plt.hist(X2,density=True, edgecolor="black",label="histogram",bins=25)
plt.plot(abs2,pdf2,label="pdf")
plt.plot(abs2,cdf2,label="cdf")
plt.legend()
plt.title("Uniforme")

plt.subplot(2,2,3)
plt.hist(X3,density=True, edgecolor="black",label="histogram",bins=25)
plt.plot(abs3,pdf3,label="pdf")
plt.plot(abs3,cdf3,label="cdf")
plt.legend()
plt.title("Triangulaire")

plt.subplot(2,2,4)
plt.hist(X4,density=True, edgecolor="black",label="histogram",bins=25)
plt.plot(abs4,pdf4,label="pdf")
plt.plot(abs4,cdf4,label="cdf")
plt.legend()
plt.title("Exponentielle")

plt.gcf().subplots_adjust( wspace = 0.5, hspace = 0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity

np.random.seed(4500)
```

```python
#Q1
N=100
X1=norm.rvs(loc=0,scale=1,size=int(0.3*N))
X2=norm.rvs(loc=5,scale=1,size=int(0.7*N))
```

```python
#Q2
Y=np.concatenate((X1,X2))
```

```python
#Q3_a
X_plot = np.linspace(-5,10,1000)
#Q3_b
true_density = norm.pdf(X_plot,scale=1,loc=0)*0.3+0.7*norm.pdf(X_plot,scale=1,loc=5)
```

```python
#Q4_a
kde = KernelDensity(kernel='gaussian', bandwidth=0.7).fit(Y[:,np.newaxis])

#Q4_b
density = np.exp(kde.score_samples(X_plot[:,np.newaxis]))

#Q4_c
# affichage : vraie densitÃ© et estimation
fig = plt.figure()
plt.fill(X_plot, true_density, fc='b', alpha=0.2, label='Vraie densitÃ©')
plt.plot(X_plot, density, '-', label="Estimation")
plt.legend(loc='upper left')
plt.show()
```
